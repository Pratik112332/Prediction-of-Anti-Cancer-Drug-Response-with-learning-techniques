```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('output/10D/seed0/lr0-01/CaDRReS_pred_1000_ckpt.csv')
df.head()
```

```python
x_names = list(df.columns.values)
%matplotlib inline
n_rows, n_cols = 0, 0
size_r, size_c = df.shape
plt.figure(figsize=(16, 12))

count = 2

labels = df['Unnamed: 0'].values[2:9]
for index, col in zip(range(size_r), df.columns.values[1:]):
    values = df[col].values[2:9]
    plt.subplot(2, 2, index + 1) 
    sns.barplot(y=labels, x=values)
    plt.xlabel(col)
    if count == 0:
        break
    count -= 1
plt.tight_layout()
plt.show()
```
